```python
import matplotlib.pyplot as plt
import numpy as np

from PIL import Image
import face_recognition
```

```python
%matplotlib inline
```

```python
image = face_recognition.load_image_file("Matrix.jpg")
```

```python
face_locations = face_recognition.face_locations(image, number_of_times_to_upsample=0, model="cnn")
```

```python
# The number of faces it found!

print("I found {} face(s) on this image.".format(len(face_locations)))
```

```python
for face_location in face_locations:

    # Print the location of each face in this image
    top, right, bottom, left = face_location
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

    # You can access the actual face itself like this:
    face_image = image[top:bottom, left:right]
    pil_image = Image.fromarray(face_image)
    plt.figure() # to make a new empty plot so it shows all the faces in one page!
    plt.imshow(pil_image)
```
